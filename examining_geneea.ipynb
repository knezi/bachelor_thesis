```python
import sys
import json
from geneea.analyzer.model import f2converter
from geneea.analyzer.model.x3 import X3
from collections import defaultdict
```

```python
with open("data/geneea_data.json", encoding='utf8') as analysisReader:
   # datafields = ['docId', 'entities', 'sentiment', 'tags', 'relations']
#, 'paragraphs', entities
    datafields = ['entities']
    textual_entities = defaultdict(int)
    for raw in map(json.loads, analysisReader):
        analysis = f2converter.fromDict(raw)
        #line = dict()
        #for x in datafields:
            #line[x] = getattr(analysis, x)
        
        #for ent in line['entities']:
        for ent in analysis.entities:
                text = ent.stdForm
                #if len(text.split(' ')) > 1:
                textual_entities[text] += 1
```
